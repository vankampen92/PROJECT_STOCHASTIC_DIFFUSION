# SIRD Model

### Types and Interpretation:

|Type | Interpretation                                     |
|:----|:---------------------------------------------------|
|$S$  | Susceptible individuals                            |
|$I$  | Infectious individuals                             |
|$R$  | Removed individuals (either recovered or deceased) |

### Rules:
For reaction constants $i, r > 0$, the following rules describe type changes:

$$
\begin{alignat*}{4}
	S + \;&I &&\xrightarrow{i}  &&I + I, &\quad&\text{[Infection]}\\
	      &I &&\xrightarrow{r}  &&R,     &\quad&\text{[Recovery]}\\
\end{alignat*}
$$

## Simulations

```python
import numpy as np
import matplotlib.pyplot as plt
import RBSE as rbse
```

## by SSA with Gillespie Algorithm

```python
def updateStates(rExc, S, I, R, D):
    if rExc == 0:         # Infection
        S -= 1
        I += 1
        return np.array(S), np.array(I), np.array(R), np.array(D)
    if rExc == 1:         # Recovery
        I -= 1
        R += 1
        return np.array(S), np.array(I), np.array(R), np.array(D)
    if rExc == 2:         # Death
        I -= 1
        D += 1
        return np.array(S), np.array(I), np.array(R), np.array(D)
    
def updatePropensities(S, I, R, D, N, i, r, d):
    kappa = np.zeros(3)
    kappa[0] = S*I/N * i # Infection
    kappa[1] =   I   * r # Recovery
    kappa[2] =   I   * d # Death
    return kappa
```

### Parameters:

```python
ny    = 1
T     = 500
N     = 10000

i     = 0.1*np.ones(ny)
r0    = 0.01
d0 = 0.005

frac  = 1.
Iinit = np.array(3)

#figsize = (7,4)
figsize = (14,8)
```

```python
s = 0
tSSA, SSSA, ISSA, RSSA, DSSA = rbse.Gillespie(T=T, N=N, Iinit=Iinit, frac=frac, seed=s,
                                              updateStates=updateStates,
                                              updatePropensities=updatePropensities,
                                              i=i, r=r0*np.ones(ny), d=d0*np.ones(ny))

fig = rbse.plotFilled(N, T, tSSA, SSSA, ISSA, RSSA, DSSA,
                      ptitle='SSA Simulation of SIR',
                      figsize=figsize,
                      plegend=False)
```

# SIRD Model with Discrete Age Classes

### Types and Interpretation:
for $\alpha \in A$  

|Type          | Interpretation          |                     |
|:-------------|:------------------------|:--------------------|
|$S^{\alpha}$  | Susceptible individuals |of subtype $\alpha$|
|$I^{\alpha}$  | Infectious individuals  |of subtype $\alpha$
|$R^{\alpha}$  | Recovered individuals   |of subtype $\alpha$|
|$D^{\alpha}$  | Deceased individuals    |of subtype $\alpha$|

We consider 3 age classes: young (0), middle (1), old(2).

### Rules:
For reaction constants $i_{\alpha,\alpha^\prime}, r_{\alpha},  d_{\alpha} > 0$, $\alpha, \alpha^\prime \in A$, the following rules describe type changes:

$$
\begin{alignat*}{4}
	S^{\alpha} + \;&I^{\alpha^\prime} &&\xrightarrow{i_{\alpha,\alpha^\prime}} \;%
                                                               &&I^{\alpha} + I^{\alpha^\prime}, &\quad \alpha, \alpha^\prime & \in  A  &\quad&\text{[Infection]}\\
	                 &I^\alpha     &&\xrightarrow{r_{\alpha}}  &&R^\alpha,  &\quad \alpha &\in A  &\quad&\text{[Recovery]}\\
                     	                 &I^\alpha     &&\xrightarrow{d_{\alpha}}  &&D^\alpha,         &\quad \alpha &\in A  &\quad&\text{[Death]}
\end{alignat*}
$$

```python
def updateStates(rExc, S, I, R, D):
    ny = len(S) # number of sub-types
    ry = 0
    if rExc < ny*ny:                          # Infection
        S[rExc      // ny] -= 1
        I[rExc      // ny] += 1
        return np.array(S), np.array(I), np.array(R), np.array(D)
    ry += ny*ny
    if rExc >= ry and rExc < ry + ny:         # Recovery
        I[rExc-ry]         -= 1
        R[rExc-ry]         += 1
        return np.array(S), np.array(I), np.array(R), np.array(D)
    ry +=ny
    if rExc >= ry and rExc < ry + ny:         # Death
        I[rExc-ry]         -= 1
        D[rExc-ry]         += 1
        return np.array(S), np.array(I), np.array(R), np.array(D)
    
def updatePropensities(S, I, R, D, N, i, r, d):
    ny = len(S) # number of sub-types
    kappa = np.zeros(ny**2 + 2*ny)
    kappa[    0     : ny**2     ] = (np.outer(S, I)/N * i).flatten()  # Infection
    kappa[ny**2     : ny**2+  ny] =              I    * r             # Recovery
    kappa[ny**2+ ny :           ] =              I    * d             # Death
    return kappa
```

### Parameters:

```python
ny    = 3
T     = 500
N     = 10000

i     = 0.1*np.ones(ny)
r0    = 0.01
d0 = 0.005

frac  = np.array([0.25, 0.5, 0.25])
Iinit = np.array([1, 1, 1])

#figsize = (7,4)
figsize = (14,8)
```

```python
s = 0
tSSA, SSSA, ISSA, RSSA, DSSA = rbse.Gillespie(T=T, N=N, Iinit=Iinit, frac=frac, seed=s,
                                              updateStates=updateStates,
                                              updatePropensities=updatePropensities,
                                              i=i, r=r0*np.ones(ny), d=d0*np.ones(ny))
fig = rbse.plotFilled(N, T, tSSA, SSSA, ISSA, RSSA, DSSA,
                      ptitle='SSA Simulation of SIR',
                      figsize=figsize,
                      plegend=False)
```

# SIRD Model with Discrete Age Classes and testing

### Types and Interpretation:
for $\alpha \in A$, $\theta \in \Theta$  

|Type          | Interpretation          |                     |
|:-------------|:------------------------|:--------------------|
|$S^{\alpha}$  | Susceptible individuals |of subtype $\alpha$|
|$I^{\alpha, \theta}$  | Infectious individuals  |of subtypes $\alpha$ and $\theta$|
|$R^{\alpha, \theta}$  | Recovered individuals   |of subtypes $\alpha$ and $\theta$|
|$D^{\alpha, \theta}$  | Deceased individuals    |of subtypes $\alpha$ and $\theta$|

We consider 3 age classes: young (0), middle (1), old(2).  
We consider 3 classes of detection: undetected(0), detected(1) - we focus only on the detection of Infected!

### Rules:
For reaction constants $i_{\alpha,\alpha^\prime,\theta^\prime}, r_{\alpha,\theta},  d_{\alpha,\theta}, \delta_{\alpha} > 0$, $\alpha, \alpha^\prime \in A$, $\theta, \theta^\prime \in \Theta$, the following rules describe type changes:

$$
\begin{alignat*}{4}
	S^{\alpha} + \;&I^{\alpha^\prime, \theta^\prime} &&\xrightarrow{i_{\alpha,\alpha^\prime,\theta^\prime}} \;%
                                                               &&I^{\alpha} + I^{\alpha^\prime,\theta^\prime}, &\quad \alpha, \alpha^\prime & \in  A, \theta^\prime \in \Theta  &\quad&\text{[Infection]}\\
	                 &I^{\alpha,\theta}     &&\xrightarrow{r_{\alpha,\theta}}  &&R^{\alpha,\theta},  &\quad \alpha &\in A, \theta \in \Theta  &\quad&\text{[Recovery]}\\
                     	                 &I^{\alpha, \theta}     &&\xrightarrow{d_{\alpha,\theta}}  &&D^{\alpha,\theta},         &\quad \alpha &\in A, \theta \in \Theta  &\quad&\text{[Death]}\\
                                                              	                 &I^{\alpha,0}     &&\xrightarrow{\delta_{\alpha}}  &&I^{\alpha,1},         &\quad \alpha &\in A &\quad&\text{[Detection of Infection]}
\end{alignat*}
$$

```python
def updateStates(rExc, S, I, R, D):
    if rExc < 9:                           # Infection by undetected Infectious
        S[rExc      // 3, 0] -= 1
        I[rExc      // 3, 0] += 1
        return np.array(S), np.array(I), np.array(R), np.array(D)
    if rExc >= 9 and rExc < 18:            # Infection by detected Infectious
        S[rExc-9    // 3, 1] -= 1
        I[rExc-9    // 3, 1] += 1
        return np.array(S), np.array(I), np.array(R), np.array(D)
    
    if rExc >= 18 and rExc < 21:           # Recovery of undetected Infected
        I[rExc-18       , 0] -= 1
        R[rExc-18       , 0] += 1
        return np.array(S), np.array(I), np.array(R), np.array(D)
    if rExc >= 21 and rExc < 24:           # Recovery of detected Infected
        I[rExc-21       , 1] -= 1
        R[rExc-21       , 1] += 1
        return np.array(S), np.array(I), np.array(R), np.array(D)

    if rExc >= 24 and rExc < 27:         # Death of undetected Infected
        I[rExc-24       , 0] -= 1
        D[rExc-24       , 0] += 1
        return np.array(S), np.array(I), np.array(R), np.array(D)
    if rExc >= 27 and rExc < 30:         # Death of detected Infected
        I[rExc-27       , 1] -= 1
        D[rExc-27       , 1] += 1
        return np.array(S), np.array(I), np.array(R), np.array(D)
    
    if rExc >= 30 and rExc < 33:         # Detection of Infected
        I[rExc-30       , 0] -= 1
        I[rExc-30       , 1] += 1
        return np.array(S), np.array(I), np.array(R), np.array(D)
    
def updatePropensities(S, I, R, D, N, i, r, d, delta):
    kappa = np.zeros(33)
    kappa[ 0: 9] = (np.outer(S[:,0], I[:,0])/N * i[:,:,0]).flatten()  # Infection by undetected Infections
    kappa[ 9:18] = (np.outer(S[:,0], I[:,1])/N * i[:,:,1]).flatten()  # Infection by   detected Infections
    kappa[18:21] =                   I[:,0]    * r[:,0]               # Recovery of  undetected Infected
    kappa[21:24] =                   I[:,1]    * r[:,1]               # Recovery of    detected Infected
    kappa[24:27] =                   I[:,0]    * d[:,0]               # Death    of    detected Infected
    kappa[27:30] =                   I[:,1]    * d[:,1]               # Death    of    detected Infected
    kappa[30:33] =                   I[:,0]    * delta                # Death    of    detected Infected
    
    return kappa
```
